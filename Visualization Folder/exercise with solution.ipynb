## Pandas Plot Exercise
Description of Dataset:

Date: It gives the date of which stocks details are given.
Open: It gives the opening price of stock on that date.
High: It gives the highest price to which the stock ascened on that day.
Low: It gives the highest price to which the stock plummeted on that day.
Close: It gives the closing price of stock on that date.
Volume: It gives the amount of stock traded on that date.
Adjusted Close: An adjusted closing price is a stock’s closing price on any given day of trading that has been amended to include any distributions and corporate actions that occurred at any time prior to the next day’s open.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('alphabet_stock_data.csv')
display(df.head())
print(df.shape)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date']) 
```

## 1. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates

```python
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')    
df2 = df.query("(Date >= @start_date) & (Date <= @end_date)")
df2 = df2.set_index('Date')
plt.figure(figsize=(8,6))
df2['Close'].plot(color='green')

plt.suptitle('Stock prices of Alphabet Inc.,\n01-04-2020 to 30-09-2020', \
                 fontsize=18, color='black')
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("$ price", fontsize=16, color='black')

plt.show()
```

## 12. Write a Pandas program to create a plot of stock price and trading volume of Alphabet Inc. between two specific dates.

```python
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')    
df2 = df.query("(Date >= @start_date) & (Date <= @end_date)")
df2=df2.set_index('Date')

gridspec_kw = dict(
    # Defines the heights of the two plots 
    # (bottom plot twice the size of the top plot)
    height_ratios=(1, 1),  
    # Zero space between axes
    hspace=0
)

# Setup the figure with 2 rows, sharing the x-axis and with 
# the gridspec_kw arguments defined above
fig, axes = plt.subplots(figsize=(12,8),
    nrows=2, ncols=1, sharex=True, 
    gridspec_kw=gridspec_kw
)
df2.plot(y='Close',color='blue',ax=axes[0])

df2.plot(y='Volume',color='red',ax=axes[1])
plt.title('\nAlphabet Inc. Trading Prices and Volume [01-04-2020 to 30-09-2020]', y=-0.40)
plt.show()
```

## 13. Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of Alphabet Inc. between two specific dates

```python
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')    
df2 = df.query("(Date >= @start_date) & (Date <= @end_date)")
df2=df2.set_index('Date')

gridspec_kw = dict(
    # Defines the heights of the two plots 
    # (bottom plot twice the size of the top plot)
    height_ratios=(1, 1,1,1,1,1),  
    # Zero space between axes
    hspace=0,
)

# Setup the figure with 2 rows, sharing the x-axis and with 
# the gridspec_kw arguments defined above
fig, axes = plt.subplots(figsize=(12,10),
    nrows=6, ncols=1, sharex=True, 
    gridspec_kw=gridspec_kw
)
df2.plot(y='Open',color='blue',ax=axes[0])

df2.plot(y='High',color='green',ax=axes[1])

df2.plot(y='Low',color='blue',ax=axes[2])

df2.plot(y='Close',color='green',ax=axes[3])

df2.plot(y='Adj Close',color='blue',ax=axes[4])

df2.plot(y='Volume',color='red',ax=axes[5])
plt.title('\n\nAlphabet Inc. Trading Prices and Volume [01-04-2020 to 30-09-2020]', y=-1.2)
```

## 14. Write a Pandas program to create a plot of adjusted closing prices, thirty days and forty days simple moving average of Alphabet Inc. between two specific dates.

```python
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')    
df2 = df.query("(Date >= @start_date) & (Date <= @end_date)")
df2 = df2.set_index('Date')
df2['SMA_30_days'] = df2['Adj Close'].rolling(window=30).mean() 
df2['SMA_40_days'] = df2['Adj Close'].rolling(window=40).mean()
plt.figure(figsize=(12,8))
df2['Adj Close'].plot(color='blue',label='Adjusted Closing Price')
df2['SMA_30_days'].plot(color='green',label='30 days simple moving average')
df2['SMA_40_days'].plot(color='red',label='40 days simple moving average')
plt.suptitle('Stock prices of Alphabet Inc.,\n01-04-2020 to 30-09-2020', \
                 fontsize=18, color='black')
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("$ price", fontsize=16, color='black')
plt.legend()
plt.show()
```

## 16. Write a Pandas program to create a plot to visualize daily percentage returns of Alphabet Inc. stock price between two specific dates.

```python
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')    
df2 = df.query("(Date >= @start_date) & (Date <= @end_date)")
df2 = df2.set_index('Date')
df2['daily return']=df2['Close'].pct_change(periods=1)
plt.figure(figsize=(8,6))
df2['daily return'].plot(color='green')

plt.suptitle('Daily percentage return of Alphabet Inc.,\n01-04-2020 to 30-09-2020', \
                 fontsize=18, color='black')
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("$ daily return", fontsize=16, color='black')

plt.show()
```

## 17. Write a Pandas program to plot the volatility over a period of time of Alphabet Inc. stock price between two specific dates

```python
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')    
df2 = df.query("(Date >= @start_date) & (Date <= @end_date)")
df2 = df2.set_index('Date')
df2['daily return']=df2['Close'].pct_change(periods=1)
df2=df2.asfreq('D', method='ffill')
df2['std']=df2['daily return'].rolling(window=30).std()
plt.figure(figsize=(8,6))
df2['std'].plot(color='green')

plt.suptitle('Volatility of Alphabet Inc.,\n01-04-2020 to 30-09-2020', \
                 fontsize=18, color='black')
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("$ daily return", fontsize=16, color='black')

plt.show()
```
